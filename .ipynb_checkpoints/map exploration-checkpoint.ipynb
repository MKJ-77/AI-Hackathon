```python
import numpy as np
import matplotlib.pyplot as plt

def dfs_explore(grid, start, region_mask, visited, heatmap):
    stack = [start]
    rows, cols = grid.shape
    dirs = [(0,1),(1,0),(0,-1),(-1,0)]

    while stack:
        r, c = stack.pop()
        if (r,c) in visited:
            continue
        if (r,c) not in region_mask:
            continue

        visited.add((r,c))
        heatmap[r][c] += 1

        for dr, dc in dirs:
            nr, nc = r+dr, c+dc
            if 0 <= nr < rows and 0 <= nc < cols:
                if grid[nr][nc] == 0 and (nr,nc) not in visited:
                    stack.append((nr,nc))

def assign_regions(grid, starts):
    rows, cols = grid.shape
    num_agents = len(starts)

    col_splits = np.array_split(range(cols), num_agents)
    regions = []

    for split in col_splits:
        mask = set()
        for r in range(rows):
            for c in split:
                mask.add((r,c))
        regions.append(mask)
    
    return regions

def visualize_heatmap(heatmap):
    plt.imshow(heatmap, cmap="hot", interpolation="nearest")
    plt.colorbar(label="Visit Frequency")
    plt.title("Exploration Efficiency Heatmap")
    plt.show()

if __name__ == "__main__":
    grid = np.zeros((10,10), dtype=int)

    starts = [(0,0), (9,9)]

    heatmap = np.zeros_like(grid)
    visited = set()

    regions = assign_regions(grid, starts)

    for agent, region in zip(starts, regions):
        dfs_explore(grid, agent, region, visited, heatmap)

    visualize_heatmap(heatmap)
```
