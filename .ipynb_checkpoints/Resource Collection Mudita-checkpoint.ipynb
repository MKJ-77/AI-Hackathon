```python
import random
import matplotlib.pyplot as plt
import numpy as np

GRID_SIZE = 20
NUM_AGENTS = 3
NUM_RESOURCES = 30

# Generate resource positions
resources = [(random.randint(0, GRID_SIZE-1), random.randint(0, GRID_SIZE-1)) 
             for _ in range(NUM_RESOURCES)]

# Shared task queue (simple list)
task_queue = resources.copy()

# Shuffle to distribute tasks randomly
random.shuffle(task_queue)


# ---------------------------------------------
# AGENT CLASS
# ---------------------------------------------
class ResourceAgent:
    def __init__(self, name, start):
        self.name = name
        self.position = start
        self.collected = []  # stores collected resource coordinates

    def collect_resource(self, resource):
        """Move to resource and mark as collected."""
        self.position = resource
        self.collected.append(resource)


# ---------------------------------------------
# CREATE AGENTS
# ---------------------------------------------
agents = []
start_positions = [(0, 0), (19, 19), (0, 19)]

for i in range(NUM_AGENTS):
    agents.append(ResourceAgent(f"Agent_{i+1}", start_positions[i]))


# ---------------------------------------------
# DISTRIBUTED TASK COLLECTION
# ---------------------------------------------
while task_queue:
    for agent in agents:
        if not task_queue:
            break
        # Each agent picks one task
        next_resource = task_queue.pop()
        agent.collect_resource(next_resource)


# ---------------------------------------------
# VISUALIZATION
# ---------------------------------------------
colors = ['red', 'blue', 'green', 'purple', 'orange']
plt.figure(figsize=(7, 7))

for i, agent in enumerate(agents):
    if agent.collected:
        x = [r[0] for r in agent.collected]
        y = [r[1] for r in agent.collected]
        plt.scatter(x, y, label=agent.name, s=80, color=colors[i])

# grid look
plt.grid(True, linestyle="--", alpha=0.4)
plt.title("Resource Collection by Cooperative Agents")
plt.legend()
plt.xlim(0, GRID_SIZE)
plt.ylim(0, GRID_SIZE)
plt.gca().invert_yaxis()  # makes grid look natural

plt.show()

# Print results
for agent in agents:
    print(f"{agent.name} collected {len(agent.collected)} resources -> {agent.collected}")
```
